```python
from bubble_sort import bubble_sort
from quick_sort import quick_sort

import random
import time
import matplotlib.pyplot as plt
```

```python
def random_array (n):
    return [random.randint(0, 10000) for _ in range(n)]


def sorted_array (n):
    return list(range(n))


def reverse_array (n):
    return list(range(n, 0, -1))

def half_sortedarray (n):
    arr = list(range(n))
    
    if n > 1:
        arr[n//2], arr[n//2 - 1] = arr[n//2 - 1], arr[n//2]
    
    return arr
```

```python
sizes = [1000, 5000, 10000, 50000]
```

```python
def test (array, name):
    
    times_bubblesort = []
    times_quicksort = []


    for n in sizes:

        arr = array(n)
        arr2 = arr.copy()

        # измерение для первого алгоритма
        start = time.time()
        bubble_sort(arr)
        end = time.time()
        times_bubblesort.append(end - start)

        # измерение для второго алгоритма
        start = time.time()
        quick_sort(arr2, 0, len(arr2) - 1)
        end = time.time()
        times_quicksort.append(end - start)


    # построение графика
    plt.plot(sizes, times_bubblesort, marker='o', label='BubbleSort function O(n²)')
    plt.plot(sizes, times_quicksort, marker='o', label='QuickSort function O(n(logn))')

    plt.xlabel("Размер входных данных (n)")
    plt.ylabel("Время выполнения (секунды)")
    plt.title(name)

    plt.legend()
    plt.grid(True)

    plt.show()
```

```python
test(random_array, "Случайный массив")
```

```python
test(sorted_array, "Отсортированный массив")
```

```python
test(reverse_array, "Обратно-отсортированный массив")
```

```python
test(half_sortedarray, "Почти отсортированный массив")
```
